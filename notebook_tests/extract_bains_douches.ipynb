```python
import pandas as pd 

df = pd.read_csv("../data/raw/ilots-de-fraicheur-equipements-activites.csv", delimiter=";")
```

```python
df.head()
```

```python
df_bains_douches = df[df["TYPE"] == "Bains-douches"]
```

```python
df_bains_douches.head()
```

```python
df_bains_douches = df_bains_douches.drop(["geo_shape", "geo_point_2d", "Bon plan fraicheur proposé par un usager (O/N)", "STATUT OUVERTURE", "Payant (O/N)"], axis=1)
df_bains_douches["arrondissement"] = df_bains_douches["Code postal"].astype(str).apply(lambda x : int(x[2:]))
```

```python
df_bains_douches
```

```python
df_bains_douches.to_csv("../data/bains_douches_paris.csv")
```

```python
columns = [column for column in df_bains_douches.columns if column.startswith("HORAIRES") == False or column == "HORAIRES PERIODE"]
horaires_columns = [column for column in df_bains_douches.columns if column not in columns]
df_bains_douches_melt = pd.melt(df_bains_douches, id_vars=columns, value_vars=horaires_columns, var_name='Jour', value_name='Horaires' )
```

```python
df_bains_douches_melt
```

```python
df_bains_douches_melt["Jour"] = df_bains_douches_melt["Jour"].apply(lambda x: x.split(" ")[1].lower())
```
