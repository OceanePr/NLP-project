```python
import pandas as pd 

df = pd.read_csv("../data/raw/les_etablissements_hospitaliers_franciliens.csv", delimiter=";")
```

```python
df
```

```python
# Extract only parisians hospitals

df_paris = df[df["dept"] == "PARIS"]
df_paris_public = df_paris[df_paris["participant_service_public_hospitalier"] == "Etablissement public de santé" ]
```

```python
# We keep only interesting columns for our project
columns = ["finess_et", "raison_sociale", "adresse_complete", "dept", "cp_ville", "num_tel", "categorie_de_l_etablissement"]
df_paris_public = df_paris_public[columns]
df_paris_public["arrondissement"] = df_paris_public["cp_ville"].astype(str).apply(lambda x: int(x.split(" ")[0][2:] if "CEDEX" not in x else int(x.split(" ")[-1])))
```

```python
df_paris_public = df_paris_public.drop(["cp_ville"], axis=1)
```

```python
df_paris_public.to_csv("../data/etablissements_hospitaliers_publis_paris.csv", index=False)
```
